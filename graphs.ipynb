# Pressure-Height, and Density-Height Graphs

This notebook will create ***Pressure-Height***, and ***Density-Height*** scatter plots based on the Barometric Formulas.

```python
#Importing Dependencies
import matplotlib.pyplot as plt
from amsimp import definitions as defs
```

## Pressure-Height Scatter Plot (Barometric Formula)

```python
plt.scatter(defs.pressure, defs.height)
plt.xlabel('Atmospheric Pressure (Pa)')
plt.ylabel('Height Above Sea Level (m)')
```

## Pressure-Density Scatter Plot (Barometric Formula)

```python
plt.scatter(defs.density, defs.height)
plt.xlabel('Atmospheric Pressure (Pa)')
plt.ylabel('Height Above Sea Level (m)')
```
